```python
import numpy as np
import pandas as pd
```

## For pancancer

```python
seeds = list(range(9))
#Generating the x-coordinates at which to evaluate the interpolated values.
all_fpr = pd.Series()
print(type(all_fpr))
all_tpr = pd.DataFrame()
for i in seeds:
    csv = pd.read_csv("../temp/fp_dnnwc"+str(i)+".csv", header = 0)
    all_fpr = pd.concat([all_fpr, csv['fpr']], ignore_index = True, sort = True)
all_fpr = pd.unique(all_fpr)
#Interpolating values of X-coordinates for 10 groups 'tpr' data. 
for i in seeds:
    csv = pd.read_csv("../temp/fp_dnnwc"+str(i)+".csv", header = 0)
    csv = csv.sort_values(by = 'fpr', ignore_index = True)
    all_tpr['seed'+str(i)] = np.interp(all_fpr, csv['fpr'], csv['tpr'])
```

```python
all_tpr.shape
```

```python
all_tpr['mean'] = all_tpr.mean(axis = 1)
all_tpr['std'] = all_tpr.std(axis =  1)
```

```python
all_tpr['all_fpr'] = all_fpr
```

```python
all_tpr['lower'] = all_tpr['mean'].sub(all_tpr['std'])
all_tpr['upper'] = all_tpr['mean'].add(all_tpr['std'])
```

```python
all_tpr.to_csv("../temp/DNNinterp4repeatedROC_pancancer.csv", header = True, index = True)
```

```python
all_tpr.iloc[100:105, 0:15]
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(csv['fpr'], csv['tpr'], color = 'green')
ax.scatter(all_fpr, all_tpr['seed0'], c='red', marker='+', linewidths = 1.0)
plt.savefig("dnninterp0_pancancer.pdf", dpi=300)
```
