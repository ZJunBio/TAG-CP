```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import numpy as np
from utils import setup_seed, metric_auc
from earlystopping import EarlyStopping
from sklearn import preprocessing
from sklearn.model_selection import cross_validate, StratifiedKFold, GridSearchCV, KFold

setup_seed(2022)
```

### Make Datasets and DataLoader

```python
train = np.load("../data/datasets/train_tensor_csigma005.npy", allow_pickle = True)
train_label = np.load("../data/datasets/train_label_csima005.npy", allow_pickle = True)
test = np.load("../data/datasets/val_tensor_csigma005.npy", allow_pickle = True)
test_label = np.load("../data/datasets/val_label_csima005.npy", allow_pickle = True)

train = np.concatenate((train, test), axis = 0)
train_label = np.concatenate((train_label, test_label), axis = -1)
print(train.shape, train_label.shape)
```

### Define classes

```python
class MakeDatasets(Dataset):
    def __init__(self, data, label):
        self.tensor = data
        self.labels = label
        self.tensor = preprocessing.normalize(self.tensor)
        self.tensor = torch.tensor(self.tensor)
        self.labels = torch.tensor(self.labels, dtype = torch.float32).reshape((len(self.labels),1))
    def __len__(self):
        return len(self.labels)
    def __getitem__(self, idx):
        vector = self.tensor[idx,:]
        label = self.labels[idx]
        return vector, label
```

```python
class Params():
    def __init__(self):
        self.dim1 = 841
        self.dim2 = 512
        self.dim3 = 256
        self.dim4 = 128
        self.dim5 = 64
        self.dim6 = 10
        self.lr = 0.001
        self.momentum = 0.9
        self.epochs = 500
        self.batchsize = 256 
        self.save_path = "../models/DNN/"
params = Params()
```

```python
# Define model
class DNN(nn.Module):
    def __init__(self, params):
        super().__init__()
        self.linear1 = nn.Linear(params.dim1, params.dim2)
        self.bn1 = nn.BatchNorm1d(params.dim2)
        self.linear2 = nn.Linear(params.dim2, params.dim3)
        self.bn2 = nn.BatchNorm1d(params.dim3)
        self.linear3 = nn.Linear(params.dim3, params.dim5)
        self.bn3 = nn.BatchNorm1d(params.dim5)
        self.linear4 = nn.Linear(params.dim5, params.dim6)
        self.bn4 = nn.BatchNorm1d(params.dim6)
        self.linear5 = nn.Linear(params.dim6, 1)
    def forward(self, x):
        x = self.linear1(x)
        x = self.bn1(x)
        x = F.relu(x)
        x = self.linear2(x)
        x = self.bn2(x)
        x = F.relu(x)
        x = self.linear3(x)
        x = self.bn3(x)
        x = F.relu(x)
        x = self.linear4(x)
        x = self.bn4(x)
        x = F.relu(x)
        x = self.linear5(x)
        x = torch.sigmoid(x)
        return x
```

```python
# Get cpu, gpu or mps device for training.
device = (
    "cuda"
    if torch.cuda.is_available()
    else "mps"
    if torch.backends.mps.is_available()
    else "cpu"
)
print(f"Using {device} device")
```

### Function of train and validation

```python
def train_loop(dataloader, model, loss_fn, optimizer):
    # Set the model to training mode - important for batch normalization and dropout layers
    # Unnecessary in this situation but added for best practices
    pred_arr = np.array([])
    label_arr = np.array([])
    train_loss = 0
    size, num_batches = len(dataloader.dataset), len(dataloader)
    for batch, (X, y) in enumerate(dataloader):
        model.train()
        # Compute prediction and loss
        pred = model(X)
        loss = loss_fn(pred, y)
        train_loss += loss

        # Backpropagation
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        
        pred_arr = np.hstack((pred_arr, np.reshape(pred.detach().numpy(), len(pred))))
        label_arr = np.hstack((label_arr, np.reshape(np.array(y), len(y))))
        
        if batch % 100 == 0:
            loss, current = loss.item(), (batch + 1) * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
    train_loss /= num_batches
    auc = metric_auc(pred_arr, label_arr)
    print(f"Train AUC: {auc:.5f}, Avg loss: {train_loss:>8f}")


def test_loop(dataloader, model, loss_fn, iter = None):
    # Set the model to evaluation mode - important for batch normalization and dropout layers
    # Unnecessary in this situation but added for best practices
    model.eval()
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0

    # Evaluating the model with torch.no_grad() ensures that no gradients are computed during test mode
    # also serves to reduce unnecessary gradient computations and memory usage for tensors with requires_grad=True
    with torch.no_grad():
        pred_arr = np.array([])
        label_arr = np.array([])
        for X, y in dataloader:
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()
            
            pred_arr = np.hstack((pred_arr, np.reshape(pred.detach().numpy(), len(pred))))
            label_arr = np.hstack((label_arr, np.reshape(np.array(y), len(y))))
            
    auc = metric_auc(pred_arr, label_arr)
    test_loss /= num_batches
    correct /= size
    print(f"Validation AUC: {auc:0.5f}, Avg loss: {test_loss:>8f} \n")
    
    return test_loss
```

### Cross validation step

```python
skf = StratifiedKFold(n_splits=9, random_state = 2023, shuffle = True)
for i, (train_index, val_index) in enumerate(skf.split(train, train_label)):
    train_data = train[train_index,:]
    train_lab = train_label[train_index]
    valid_data = train[val_index,:]
    valid_lab = train_label[val_index]
    
    train_data = MakeDatasets(train_data, train_lab)
    test_data = MakeDatasets(valid_data, valid_lab)
    
    train_loader = DataLoader(train_data, batch_size = params.batchsize, shuffle = True)
    test_loader = DataLoader(test_data, batch_size = params.batchsize, shuffle = True)
    print(len(train_loader), len(test_loader))
    
    model = DNN(params).to(device)
    print(model)
    
    estop = EarlyStopping(patience = 15)
    loss_fn = nn.BCELoss()
    optimizer = torch.optim.SGD(model.parameters(), lr = params.lr, momentum = params.momentum)
    #optimizer = torch.optim.Adam(model.parameters(), lr=0.005)
    
    epochs = params.epochs
    for t in range(epochs):
        print(f"Epoch {t+1}\n-------------------------------")
        train_loop(train_loader, model, loss_fn, optimizer)
        test_loss = test_loop(test_loader, model, loss_fn)
        
        #early stop mechanism
        estop(test_loss)
        if estop.earlystop:
            torch.save(model, params.save_path+"pytorch_es" + str(i) + ".pt")
            break
    print("Done!")
```
